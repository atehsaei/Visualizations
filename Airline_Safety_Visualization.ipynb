```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/airline-safety/airline-safety.csv")
```

```python
df.head()
```

```python
sns.regplot(y=df['fatal_accidents_85_99'], x=df['fatalities_85_99'])
```

# No correlation between fatal accidents and fatalities between yrs. 1985-99

```python
sns.regplot(y=df['fatal_accidents_00_14'], x=df['fatalities_00_14'])
```

# No correlation between fatal accidents and fatalities between yrs. 2000-14

# 8 Air lines with most fatalities 

```python
Ft_airln=df[(df['airline']=='Air France')|(df['airline']=='Air India*') | (df['airline']=='American*') |\
            (df['airline']=='China Airlines') | (df['airline']=='Gulf Air') |\
            (df['airline']=='Gulf Air')| (df['airline']=='Malaysia Airlines')| (df['airline']=='TAM')|\
            (df['airline']=='Kenya Airways')]
```

```python
Ft_airln
```

# Kernel Density Estimate (KDE) plot for Fatalities between yrs. 1985-2014

```python
sns.kdeplot(Ft_airln['fatalities_00_14'],Ft_airln['fatalities_85_99'], shade=True, shade_lowest=True,\
           cmap='inferno')
sns.kdeplot(Ft_airln['fatalities_00_14'],Ft_airln['fatalities_85_99'],cmap='cool')
```

# Kernel Density Estimate (KDE) plot for Fatal accidents
                                 between yrs. 1985-2014

```python
sns.kdeplot(Ft_airln['fatal_accidents_00_14'],Ft_airln['fatal_accidents_85_99'], shade=True, shade_lowest=True,\
           cmap='inferno')
sns.kdeplot(Ft_airln['fatal_accidents_00_14'],Ft_airln['fatal_accidents_85_99'],cmap='cool')
```

# Kernel Density Estimate (KDE) plot for incidents between yrs. 1985-2014

```python
sns.kdeplot(Ft_airln['incidents_00_14'],Ft_airln['incidents_85_99'], fill=True, thresh=0,levels=100, cmap="mako")
```
